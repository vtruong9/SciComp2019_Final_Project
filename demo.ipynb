```python
# find motifxofxinterest in Crick strand

bases = {'A':'T','T':'A','C':'G','G':'C'}

def reverse_complement(r):
    output = []
    for i in range(len(r)):
        output.append(bases[r[i]])
#    return (("".join(output))[::-1])
    return ((output)[::-1])
motifrc = reverse_complement(motif)
motifrca = np.array(motifrc)
print(motifrca)
```

```python
# Roman numeral converter
#https://www.w3resource.com/python-exercises/class-exercises/python-class-exercise-1.php
def int_to_roman(num):
        val = [
            1000, 900, 500, 400,
            100, 90, 50, 40,
            10, 9, 5, 4,
            1
            ]
        syb = [
            "M", "CM", "D", "CD",
            "C", "XC", "L", "XL",
            "X", "IX", "V", "IV",
            "I"
            ]
        roman_num = ''
        i = 0
        while  num > 0:
            for _ in range(num // val[i]):
                roman_num += syb[i]
                num -= val[i]
            i += 1
        return roman_num
```
